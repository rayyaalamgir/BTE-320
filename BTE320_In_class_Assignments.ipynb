Income Statement using Simple Income Statement Calculation:

```python
sales = float(input("enter sales ($)"))
tax_rate = float(input("Enter tax rate (%): "))

cogs = 0.4 * sales
gross_profit = sales - cogs
taxes = gross_profit * tax_rate / 100
net = gross_profit - taxes

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")
print(f"Taxes: {taxes}$")
print(f"Net Income: {net}$")
```

```python
# Calculate EOQ

#Inputs: demand, reorder cost, holding cost
demand = float(input("Enter the projected demand (units/year): "))
reorderCost = float(input("Enter the reorder cost ($/order): "))
holdingCost = float(input("Enter the holding cost ($/year/unit): "))

#Process: calculate EOQ
eoq = (2 * demand * reorderCost / holdingCost) ** 0.5

#Output: display EOQ value
print(f"Economic Order Quantity: {eoq}")
```

Rock-Paper-Scissors

```python
import random

play = input("Do you want to play a game (y/n)?")

while play == 'y':
  player = input("Choose b/w rock, paper, scissors: ").lower()
  computer = random.choice(['rock', 'paper', 'scissors'])

  print(f"Player: {player}\nComputer: {computer}")

  if player == computer:
    print("Draw")
  elif (
    (player == 'rock' and computer == 'scissors') or
    (player == 'scissors' and computer == 'paper') or
    (player == 'paper' and computer == 'rock')
    ):
    print("Player wins!")
  else:
    print("Computer wins!")

  play = input("Do you want to play again (y/n)? ")
```

Recursive Function: String Reversal

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    #return s[-1] + reverse(s[:-1])
    #return reverse(s[1:]) + s[0]
    return s[-1] + reverse(s[1:-1]) + s[0]

reverse('12345')
```

ROI Calculator

```python
def roi(cost, revenue):
  assert cost != 0
  assert isinstance(cost, (int, float))
  assert isinstance(revenue, (int, float))
  return (revenue - cost) / cost

cost = float(input("Enter cost value: "))
revenue = float(input("Enter revenue value: "))

print(roi(cost, revenue))
```

Loan Payment Calculator

```python
def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float)) and isinstance(annual_rate, (float)) and isinstance(months, (int,float))
  try:
    monthly_rate = annual_rate / 12
    if monthly_rate != 0:
      monthly_payment = monthly_rate * principal / (1 - (1 + monthly_rate)** (-months))
    else:
      monthly_payment = principal / months
  except:
      raise ValueError("Invalid input: Function inputs must be positive")
  return monthly_payment

principal = float(input("Enter the loan amount: "))
annual_rate = float(input("Enter the annual interest rate: "))
months = float(input("Enter the number of months:"))
print (round(loan_payment(principal, annual_rate, months)))
```

Student Class

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newName):
    assert isinstance(newName, str)
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(f"Student {self.__name} is already enrolled in {newCourse}.")

s = Student("Jane Doe", 12345)
print(f"Student name: {s.get_name()}")
print(f"Classes already enrolled in: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Total number of classes enrolled: {len(s)}")
```

Classes Point and Rectangle

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(4,0)
    self.p3 = Point(4,3)
    self.p4 = Point(0,3)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p4)

  def area(self):
    return self.width * self.height
```

```python
z = Rectangle()
print(z.area())
```

Class Employee - Manager

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__hourly_wage = wage
    self.hours = hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self, newWage):
    if newWage <= 0:
      print("Invalid wage value.")
    else:
      self.__hourly_wage = newWage

  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def __str__(self):
    return f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}$"

class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus # decimal format, e.g., 0.2 for 20%

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly payment is {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly payment is {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus.")
```

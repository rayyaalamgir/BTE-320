```python
sales = float(input("enter sales ($)"))
tax_rate = float(input("Enter tax rate (%): "))

cogs = 0.4 * sales
gross_profit = sales - cogs
taxes = gross_profit * tax_rate / 100
net = gross_profit - taxes

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")
print(f"Taxes: {taxes}$")
print(f"Net Income: {net}$")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operator (e.g., +, *, etc.): ")

res = None

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

  print(f"Result: {res}")
```

```python
print("Hour", "Fee")
hour = 1
while hour < 9:
  fee = 5 + 2.5 * hour
  if fee < 10.0:
    fee = 10.0
  elif fee > 20.0:
    fee = 20.0
  print(f"{hour}   {fee}")
  hour += 1
```

```python
import random

play = input("Do you want to play a game (y/n)?")

while play == 'y':
  player = input("Choose b/w rock, paper, scissors: ").lower()
  computer = random.choice(['rock', 'paper', 'scissors'])

  print(f"Player: {player}\nComputer: {computer}")

if player == computer:
  print("Draw")
elif (
    (player == 'rock' and computer == 'scissors') or
    (player == 'scissors' and computer == 'paper') or
    (player == 'paper' and computer == 'rock')
    ):
    print("Player wins!")
else:
  print("Computer wins!")

play = input("Do you want to play again (y/n)? ")
```
